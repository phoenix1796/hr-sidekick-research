```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from process import getDataset

hr = getDataset('HR_comma_sep.csv')
d1 = pd.DataFrame(hr['data'],columns = hr['headers']).convert_objects(convert_numeric=True)


# Importing the dataset
# d1 = pd.read_csv('HR.csv')

X = d1.iloc[:,[0,1,2,3,4,5,7]].values
y = d1.iloc[:,6].values 

```

```python
pd.crosstab(d1.last_evaluation,d1.left).plot(kind='bar')
plt.title('Job Left Frequency for last_evaluation')
plt.xlabel('last_evaluation')
plt.ylabel('Frequency of Job Left')
```

```python
ticks = np.arange(min(d1.loc[:,'satisfaction_level']), max(d1.loc[:,'satisfaction_level'])+1, 0.4)
pd.crosstab(d1.satisfaction_level,d1.left).plot(xticks=ticks ,kind='bar')
plt.title('Job Left Frequency for satisfaction_level')
plt.xlabel('satisfaction_level')
plt.ylabel('Frequency of Job Left')
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

# Fitting Logistic Regression to the Training set
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train) 

# Predicting the Test set results
y_pred = classifier.predict(X_test) 

# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```
